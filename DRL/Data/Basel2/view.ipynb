```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pickle
import random
import time,os
from tqdm import tqdm
```

```python
filename = 'measurement_29.pkl'
infile = open(filename,'rb')
data_set = pickle.load(infile)
infile.close()
```

```python
image = data_set['chan0']['data']
image = np.rot90(np.rot90(image))
np.save('basel2_29',image)
plt.imshow(image)
plt.colorbar()
plt.show()
```
