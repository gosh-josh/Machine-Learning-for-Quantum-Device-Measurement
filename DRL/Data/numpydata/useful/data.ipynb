```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pickle
import random
import time,os
from tqdm import tqdm
```

```python
cwd = os.getcwd()
for file in tqdm(os.listdir(cwd)):
    filename = os.fsdecode(file)
    print(filename)
    if filename.endswith(".npy"):
        data = np.load(filename)
        print(np.shape(data))
        plt.imshow(data)
        plt.colorbar()
        plt.show()
```
